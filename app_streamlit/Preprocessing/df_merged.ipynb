```python
import pandas as pd
from functools import reduce

load_path_1 = r'C:\dev\clinicaltrials-study\data\processed\studies_clean.csv'
load_path_2 = r'C:\dev\clinicaltrials-study\data\processed\interventions_clean.csv'
load_path_3 = r'C:\dev\clinicaltrials-study\data\processed\conditions_clean.csv'
load_path_4 = r'C:\dev\clinicaltrials-study\data\processed\designs_clean.csv'
load_path_5 = r'C:\dev\clinicaltrials-study\data\processed\eligibilities_clean.csv'
load_path_6 = r'C:\dev\clinicaltrials-study\data\processed\sponsor_clean.csv'

df_studies = pd.read_csv(load_path_1)
df_interventions = pd.read_csv(load_path_2)
df_conditions = pd.read_csv(load_path_3)
df_designs = pd.read_csv(load_path_4)
df_eligibilities = pd.read_csv(load_path_5)
df_sponsors = pd.read_csv(load_path_6)
```

```python
# 모든 전처리 데이터셋 통합(reduce)
dfs_to_merge = [
    df_studies,
    df_interventions,
    df_conditions,
    df_designs,
    df_eligibilities,
    df_sponsors
]

df_merged = reduce(lambda left, right: pd.merge(left, right, on='nct_id', how='left'), dfs_to_merge)
```

```python
# 결측치 있는 컬럼만 0으로 채우기 
target_counts = [5922, 1]
flag_cols = [col for col in df_merged.columns if df_merged[col].isnull().sum() in target_counts]
df_merged[flag_cols] = df_merged[flag_cols].fillna(0)
print(f'처리된 컬럼 개수: {len(flag_cols)}')
```

```python
# 최종 확인
print("Total missing values in the dataset:", df_merged.isnull().sum().sum())
print("Final shape after merging and filling nulls:", df_merged.shape)
```

```python
# 최종 데이터셋 저장
import os
target_path = r'C:\dev\clinicaltrials-study\data\processed'
file_name = 'df_merged.csv'
full_save_path = os.path.join(target_path, file_name)
df_merged.to_csv(full_save_path, index=False)
```
